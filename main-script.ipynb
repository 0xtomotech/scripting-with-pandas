### Setting values

```python
coffee.loc[1, "Units Sold"] = 10
coffee.head()
```

```python
# optmized way to update values
coffee.at[1, "Units Sold"] = 20
coffee.head()
```

### Sort Values

```python
coffee.sort_values(['Units Sold', 'Coffee Type'], ascending=[0,1])
```

```python
# Iterate over dataframe with for loop
for index, row in coffee.iterrows():
    print(index)
    print(row)
    print("Coffee Type of Row:", row["Coffee Type"])
```
